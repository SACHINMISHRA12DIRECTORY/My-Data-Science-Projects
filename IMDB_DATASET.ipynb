```python
# Create a genre-level report with metrics average rating, the average number of votes, 
# and average revenue. What is the average rating of the ‘Horror’ genre?
```

```python
df['Rating'].mean()
df['Votes'].mean()
df['Revenue_millions'].mean()
df['genre-level report'] = df['Rating'].mean() + df['Votes'].mean() + df['Revenue_millions'].mean()
```

```python
df.head()
```

```python
import pandas as pd

f = pd.DataFrame(df)

# Group the data by genre and calculate metrics
genre_metrics = df.groupby('Genre').agg({
    'Rating': 'mean',            # Average rating
    'Votes': 'mean',            # Average number of votes
    'Revenue_millions': 'mean'  # Average revenue
}).reset_index()

# Display the genre-level report
print(genre_metrics)

```

```python
# Find the average rating for the 'Horror' genre
average_rating_horror = genre_metrics[genre_metrics['Genre'] == 'Horror']['Rating'].values[0]
average_rating_horror
```

```python
# Create Custom column (Runtime_hours) using Runtime_miuntes and check which genre has the longest  Runtime_hours.
df['Runtime_miuntes']
```

```python


# Step 1: Create the 'Runtime_hours' column (convert minutes to hours)
df['Runtime_hours'] = df['Runtime_minutes'] / 60

# Step 2: Group the data by genre and calculate the maximum 'Runtime_hours' for each genre
genre_max_runtime = df.groupby('Genre')['Runtime_hours'].max().reset_index()

# Step 3: Identify the genre with the longest runtime
longest_runtime_genre = genre_max_runtime[genre_max_runtime['Runtime_hours'] == genre_max_runtime['Runtime_hours'].max()]

# Display the genre with the longest runtime
print("Genre with the Longest Runtime (in hours):")
print(longest_runtime_genre)
```

```python
# Find the revenue of Genre (Action, Adventure, Horror) released before year 2014
```

```python

# Define the genres and the cutoff year
genres_to_include = ['Action', 'Adventure', 'Horror']
cutoff_year = 2014

# Step 1: Filter the DataFrame based on genres and release year
filtered_df = df[(df['Genre'].isin(genres_to_include)) & (df['Year'] < cutoff_year)]

# Step 2: Calculate the total revenue for each genre
genre_revenue_totals = filtered_df.groupby('Genre')['Revenue_millions'].sum().reset_index()

# Display the total revenue for each genre
print("Total Revenue for Genres (Action, Adventure, Horror) Released Before 2014:")
print(genre_revenue_totals)
```

```python


# Define the year to filter
target_year = 2015

# Step 1: Filter the DataFrame based on the year
filtered_df = df[df['Year'] == target_year]

# Step 2: Calculate the total revenue for the filtered year
total_revenue_2015 = filtered_df['Revenue_millions'].sum()

# Display the total revenue for the year 2015
print(f"Total revenue generated by movies in the year 2015: {total_revenue_2015} million")
```

```python

# Define the target genre and year
target_genre = 'Adventure'
target_year = 2015

# Step 1: Filter the DataFrame based on genre and year
filtered_df = df[(df['Genre'] == target_genre) & (df['Year'] == target_year)]

# Step 2: Calculate the average rating for the filtered genre and year
average_rating_adventure_2015 = filtered_df['Rating'].mean()

# Display the average rating for 'Adventure' genre in the year 2015
average_rating_adventure_2015
```

```python
# Define the revenue and rating thresholds
revenue_threshold = 50
rating_threshold = 7.0

# Step 1: Filter the DataFrame based on revenue and rating
filtered_df = df[(df['Revenue_millions'] > revenue_threshold) & (df['Rating'] < rating_threshold)]

# Step 2: Count the number of movies that meet the criteria
num_movies = len(filtered_df)

# Display the count of movies that meet the criteria
print(f"Number of movies with revenue > {revenue_threshold} million and rating < {rating_threshold}: {num_movies}")
```
