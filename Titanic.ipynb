loc (label-based indexing):

    loc is primarily used for label-based indexing, which means you select data based on the labels of rows and columns.
    It uses square brackets and accepts row and column labels as input. You can use labels to specify both rows and columns.
    The syntax is df.loc[row_labels, column_labels] for DataFrames and series.loc[row_labels] for Series.
    EXAMPLE
    df.loc[2, 'Column1']  # Selects the value in row with label 2 and column 'Column1'
    df.loc[[1, 3], ['Column2', 'Column3']]  # Selects rows 1 and 3 and columns 'Column2' and 'Column3'
    df.loc[df['Column4'] > 10]  # Selects rows where 'Column4' values are greater than 10

    
    
    iloc (integer-based indexing):

    iloc is used for integer-based indexing, which means you select data using integer positions or indices.
    It also uses square brackets and accepts integer-based row and column indices.
    The syntax is df.iloc[row_indices, column_indices] for DataFrames and series.iloc[row_indices] for Series.
    EXAMPLE
    df.iloc[0, 1]  # Selects the value at row index 0 and column index 1
    df.iloc[1:3, 2:4]  # Selects rows 1 and 2 and columns 2 and 3 (exclusive of stop indices)
    df.iloc[[0, 2], [1, 3]]  # Selects rows 0 and 2 and columns 1 and 3


```python
#loc and iloc function (Index lock function)
#Syntax df.loc[strt_index:stop_index(Rows):stepsize,column_name]
#Syntax df.iloc[strt_index:stop_index(Rows):stepsize,strt_index:stop_index(Columns):stepsize]
df.loc[50:100, ['Name','Age','Fare']]
df.iloc[50:100,4:9]
df.loc[250:300,['PassengerId','Pclass','Age']]
df.iloc[500:525,1:6:2]
```

```python
#Fill the area between two plots
x = np.array([2,4,6,8])
y = x * 2
plt.plot(x,y,'-')
x1 = [2,4,6,8]
y1 = [3,5,7,9]
plt.plot(x1,y1, '-*')
plt.xlabel('X-array')
plt.ylabel('Y-array')
plt.title('Line Plot with Multiple plots on the same axis')
plt.fill_between(x,y,y1, color = 'green', alpha = 0.5)
plt.show()
```
