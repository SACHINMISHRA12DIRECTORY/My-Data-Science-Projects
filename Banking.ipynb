```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('BankChurners.csv')
```

```python
df.columns
```

```python
# Display a sample of five rows of the data frame
print("Sample of five rows of the data frame:")
print(df.head())
```

```python
# Check the shape of the data
print("\nShape of the data (number of rows and columns):")
print(df.shape)
```

```python
# Check the percentage of missing values in each column
print("\nPercentage of missing values in each column:")
print(df.isnull().mean() * 100)
```

```python
# Check if there are any duplicate rows
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
# Check the distribution of the Customer_Age column
print("\nDistribution of Customer_Age:")
print(df['Customer_Age'].describe())
```

```python
# Plot a Box-plot of Total_Revolving_Bal and Card_Category by characterizing with Attrition_Flag
plt.figure(figsize=(12, 6))
sns.boxplot(x='Card_Category', y='Total_Revolving_Bal', hue='Attrition_Flag', data=df)
plt.title("Box-plot of Total_Revolving_Bal and Card_Category by Attrition_Flag")
plt.show()
```

```python
# Plot 2 box plots and 2 pie charts for 'Credit_Limit' and 'Total_Trans_Amt'
plt.figure(figsize=(15, 8))

# Box Plot for 'Credit_Limit' by Attrition_Flag
plt.subplot(2, 2, 1)
sns.boxplot(x='Attrition_Flag', y='Credit_Limit', data=df)
plt.title("Box Plot for Credit_Limit by Attrition_Flag")

# Box Plot for 'Total_Trans_Amt' by Attrition_Flag
plt.subplot(2, 2, 2)
sns.boxplot(x='Attrition_Flag', y='Total_Trans_Amt', data=df)
plt.title("Box Plot for Total_Trans_Amt by Attrition_Flag")

# Pie Chart for 'Education_Level'
plt.subplot(2, 2, 3)
df['Education_Level'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart for Education_Level")

# Pie Chart for 'Card_Category'
plt.subplot(2, 2, 4)
df['Card_Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Pie Chart for Card_Category")

plt.show()

```

    Box Plot for Credit_Limit by Attrition_Flag:
        For existing customers (Attrition_Flag = 0), the median credit limit is relatively higher compared to attrited customers (Attrition_Flag = 1).
        There are some outliers for both existing and attrited customers, suggesting that some individuals have exceptionally high or low credit limits.
        The overall spread of credit limits is wider for existing customers, indicating a diverse range of credit limits.

    Box Plot for Total_Trans_Amt by Attrition_Flag:
        The median total transaction amount is slightly higher for existing customers compared to attrited customers.
        Both groups have a similar spread, but attrited customers generally have lower total transaction amounts.
        There are some outliers for both categories, indicating customers with exceptionally high or low total transaction amounts.

    Pie Chart for Education_Level:
        The pie chart illustrates the distribution of customers across different education levels.
        The majority of customers have completed a college education, followed by high school and post-graduate education.
        The distribution does not seem to vary significantly between existing and attrited customers, suggesting education level alone may not be a strong indicator of attrition.

    Pie Chart for Card_Category:
        The pie chart shows the distribution of customers based on their card categories.
        Most customers have a 'Blue' card, with 'Silver' being the next most common category.
        The distribution remains consistent for both existing and attrited customers, indicating that card category alone may not be a decisive factor in customer attrition.

```python
# Plot a Box-plot of Total_Revolving_Bal and Card_Category by characterizing with Attrition_Flag
plt.figure(figsize=(12, 8))
sns.boxplot(x='Card_Category', y='Total_Revolving_Bal', hue='Attrition_Flag', data=df)
plt.title("Box-plot of Total_Revolving_Bal and Card_Category by Attrition_Flag")
plt.xlabel("Card_Category")
plt.ylabel("Total_Revolving_Bal")
plt.show()
```

the visualization indicates that, particularly for Blue cards, there is a noticeable difference in the distribution of Total_Revolving_Bal between existing and attrited customers.
While card category alone may not be a decisive factor, the combination of card category and Total_Revolving_Bal can provide insights into potential patterns associated with customer attrition.

```python

# Plot a percentage segment bar graph between Education_Level and Attrition_Flag
plt.figure(figsize=(10, 6))
edu_vs_attrition = pd.crosstab(df['Education_Level'], df['Attrition_Flag'], normalize='index')
edu_vs_attrition.plot(kind='bar', stacked=True)
plt.title("Percentage Segment Bar Graph: Education_Level vs. Attrition_Flag")
plt.xlabel("Education_Level")
plt.ylabel("Percentage")
plt.show()
```

```python
# Plot a percentage segment bar graph between Income_Category and Attrition_Flag
plt.figure(figsize=(10, 6))
income_vs_attrition = pd.crosstab(df['Income_Category'], df['Attrition_Flag'], normalize='index')
income_vs_attrition.plot(kind='bar', stacked=True)
plt.title("Percentage Segment Bar Graph: Income_Category vs. Attrition_Flag")
plt.xlabel("Income_Category")
plt.ylabel("Percentage")
plt.show()
```

```python
# Drop CLIENTNUM column
df = df.drop(columns=['CLIENTNUM'])
```

```python
# Make a sub data frame with numerical columns and Attrition_Flag
numerical_df = df.select_dtypes(include=['int64', 'float64'])
numerical_df['Attrition_Flag'] = df['Attrition_Flag']
```

```python
# Plot a heatmap to view the correlation
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Plot a boxplot for the Credit_Limit column
plt.figure(figsize=(8, 6))
sns.boxplot(x='Credit_Limit', data=df)
plt.title("Boxplot for Credit_Limit")
plt.show()
```

```python
# Check for outliers in the Credit_Limit column
outliers = df[df['Credit_Limit'] > df['Credit_Limit'].quantile(0.75) + 1.5 * (df['Credit_Limit'].quantile(0.75) - df['Credit_Limit'].quantile(0.25))]
print("\nNumber of outliers in Credit_Limit:", len(outliers))
```

```python
# Map the Attrition_Flag values to 0 and 1, and standardize the columns
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
scaler = StandardScaler()
df[numerical_df.columns] = scaler.fit_transform(df[numerical_df.columns])
```
